```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('HeartDisease.csv')
df
```

```python
df['cp'] = df['cp'].astype('string')
df['thal'] = df['thal'].astype('string')
df['AHD(target)'] = df['AHD(target)'].astype('string')
```

```python
df.describe
```

```python
df.dtypes
```

```python
df['cp'] = df.cp.map({'typical': 0, 'asymptomatic': 1 ,'nonanginal': 2, 'nontypical': 3})
df['thal'] = df.thal.map({'normal': 3, 'fixed': 6 ,'reversable': 7})
df['AHD(target)'] = df['AHD(target)'].map({'No': 0, 'Yes': 1 })
df.head()
```

```python
df.dtypes
```

```python
df = df.dropna()
df
```

```python
X = np.asarray(df[['Age', 'Sex', 'cp', 'trestbps', 'chol', 'fbs','restecg' , 'exang' , 'oldpeak' ,'slope' , 'ca' , 'thal' ,' thalach' ]])
X[0:5]
```

```python
y = np.asarray(df['AHD(target)'])
y [0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
nb.fit(X_train,y_train)
```

```python
print("Naive Bayes score: ",nb.score(X_test, y_test))
```
